```python
import numpy as np
import pandas as pd
```

```python
data =  pd.read_csv('social_media_dataset.csv') ##Import csv files to colab
data.head(15)
```

```python
data.groupby('Platform').sum().head()
```

```python
total_data = data.groupby('Platform').mean() ##Group by platform
total_data.head()
```

```python
criteria = ['User Engagement','Reach and Popularity', 'Trustworthiness', 'Ease of Sharing']
bobot = [0.4,0.3,0.2,0.1]
```

```python
def norm(total_data):
  for i in criteria:
    total_data[i] = total_data[i]/total_data[i].max()

  return total_data
norm_data = norm(total_data)
norm_data.head()
```

```python
def saw(total_data,bobot):
  total_data = norm(total_data)
  hasil = total_data.mul(bobot,axis=1).sum(axis=1)
  return hasil
```

```python
hasil_saw = saw(total_data,bobot)
print("Hasil SAW =")
for platform, nilai in hasil_saw.items():
    print(f"{platform}: {nilai:.6f}")
print (f'Sosial Media terbaik = {hasil_saw.idxmax()}')
```
